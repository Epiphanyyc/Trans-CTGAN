```python
import pandas as pd

# 定义列名
columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'
]

# 定义工作类别映射
workclass_map = {
    'Private': 0, 'Self-emp-not-inc': 1, 'Self-emp-inc': 2, 'Federal-gov': 3,
    'Local-gov': 4, 'State-gov': 5, 'Without-pay': 6, 'Never-worked': 7
}

# 定义教育映射
education_map = {
    'Bachelors': 0, 'Some-college': 1, '11th': 2, 'HS-grad': 3, 'Prof-school': 4,
    'Assoc-acdm': 5, 'Assoc-voc': 6, '9th': 7, '7th-8th': 8, '12th': 9,
    'Masters': 10, '1st-4th': 11, '10th': 12, 'Doctorate': 13, '5th-6th': 14,
    'Preschool': 15
}

# 定义婚姻状况映射
marital_status_map = {
    'Married-civ-spouse': 0, 'Divorced': 1, 'Never-married': 2, 'Separated': 3,
    'Widowed': 4, 'Married-spouse-absent': 5, 'Married-AF-spouse': 6
}

# 定义职业映射
occupation_map = {
    'Tech-support': 0, 'Craft-repair': 1, 'Other-service': 2, 'Sales': 3,
    'Exec-managerial': 4, 'Prof-specialty': 5, 'Handlers-cleaners': 6,
    'Machine-op-inspct': 7, 'Adm-clerical': 8, 'Farming-fishing': 9,
    'Transport-moving': 10, 'Priv-house-serv': 11, 'Protective-serv': 12,
    'Armed-Forces': 13
}

# 定义关系映射
relationship_map = {
    'Wife': 0, 'Own-child': 1, 'Husband': 2, 'Not-in-family': 3,
    'Other-relative': 4, 'Unmarried': 5
}

# 定义种族映射
race_map = {
    'White': 0, 'Asian-Pac-Islander': 1, 'Amer-Indian-Eskimo': 2,
    'Other': 3, 'Black': 4
}

# 定义性别映射
sex_map = {'Female': 0, 'Male': 1}

# 定义国家映射
native_country_map = {
    'United-States': 0, 'Cambodia': 1, 'England': 2, 'Puerto-Rico': 3,
    'Canada': 4, 'Germany': 5, 'Outlying-US(Guam-USVI-etc)': 6, 'India': 7,
    'Japan': 8, 'Greece': 9, 'South': 10, 'China': 11, 'Cuba': 12,
    'Iran': 13, 'Honduras': 14, 'Philippines': 15, 'Italy': 16, 'Poland': 17,
    'Jamaica': 18, 'Vietnam': 19, 'Mexico': 20, 'Portugal': 21, 'Ireland': 22,
    'France': 23, 'Dominican-Republic': 24, 'Laos': 25, 'Ecuador': 26,
    'Taiwan': 27, 'Haiti': 28, 'Columbia': 29, 'Hungary': 30,
    'Guatemala': 31, 'Nicaragua': 32, 'Scotland': 33, 'Thailand': 34,
    'Yugoslavia': 35, 'El-Salvador': 36, 'Trinadad&Tobago': 37, 'Peru': 38,
    'Hong': 39, 'Holand-Netherlands': 40
}

# 定义收入映射
income_map = {'<=50K': 0, '>50K': 1}

# 读取数据
data = pd.read_csv(r'G:\DataSets\adult.data', names=columns, na_values=' ?', skipinitialspace=True)

# 数据预处理
data['workclass'] = data['workclass'].map(workclass_map).fillna(0)
data['education'] = data['education'].map(education_map).fillna(0)
data['marital-status'] = data['marital-status'].map(marital_status_map).fillna(0)
data['occupation'] = data['occupation'].map(occupation_map).fillna(0)
data['relationship'] = data['relationship'].map(relationship_map).fillna(0)
data['race'] = data['race'].map(race_map).fillna(0)
data['sex'] = data['sex'].map(sex_map)
data['native-country'] = data['native-country'].map(native_country_map).fillna(0)
data['income'] = data['income'].map(income_map)

# 保存为 CSV 文件
data.to_csv(r'G:\DataSets\adult_processed_0.csv', index=False)

print("数据预处理完成并已保存为 G:\DataSets\adult_processed.csv")
```
