```python
import numpy as np
import pandas as pd

def preprocess_data(real_path, fake_path):

    real_data = pd.read_csv(real_path).select_dtypes(include=[np.number])
    gen_data = pd.read_csv(fake_path).select_dtypes(include=[np.number])
    

    real_data = real_data.dropna()
    gen_data = gen_data.dropna()
    

    common_cols = real_data.columns.intersection(gen_data.columns)
    real_data = real_data[common_cols]
    gen_data = gen_data[common_cols]

    real_data = real_data.values
    gen_data = gen_data.values

    std_real = np.std(real_data, axis=0)
    zero_std_cols = np.where(std_real == 0)[0]
    real_data = np.delete(real_data, zero_std_cols, axis=1)
    gen_data = np.delete(gen_data, zero_std_cols, axis=1)
    
    return real_data, gen_data

def l2_distance_corr_matrix(real_data, gen_data):

    real_corr = np.corrcoef(real_data, rowvar=False)
    gen_corr = np.corrcoef(gen_data, rowvar=False)
    
    if np.isnan(real_corr).any() or np.isnan(gen_corr).any():
        raise ValueError("nan")
    
    l2_dist = np.linalg.norm(real_corr - gen_corr, ord='fro')
    return l2_dist

# real_path = '../CTGAN-main/CTGAN-main/examples/csv/train_clean.csv'
# fake_path = "../synthcity-main/tutorials/tit1.csv"
fake_path = "G:/DataSets/Fake_Dataset/CTABGAN-Cover30k1.csv"
real_path = "G:/DataSets/Churn.csv"
try:
    real_data, gen_data = preprocess_data(real_path, fake_path)
    distance = l2_distance_corr_matrix(real_data, gen_data)
    print(f"L2: {distance:.4f}")
except ValueError as e:
    print(f": {e}")
```
